```python
import os 
import pandas as pd
```

```python
nazwy = []
```

```python
max_f = len(os.listdir(r'data_scraping\07.06.2024'))
for index, file in enumerate(os.listdir(r'data_scraping\07.06.2024')):
    df = pd.read_csv(r'data_scraping\07.06.2024\\'+file)
    t = list(df['item_1'].unique())
    for x in t:
        if x not in nazwy:
            nazwy.append(x)
    t = list(df['item_2'].unique())
    for x in t:
        if x not in nazwy:
            nazwy.append(x)
    t = list(df['item_3'].unique())
    for x in t:
        if x not in nazwy:
            nazwy.append(x)
    print(f'{index}/{max_f}')
```

```python
print(sorted(nazwy))
```
